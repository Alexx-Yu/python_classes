### Сортировка с фильтрацией

```python
# Восстановить хронологию продаж таунхаусов, у которых площадь участка существенно больше площади здания.
# Объекты, проданные в один и тот же день, необходимо сортировать по значению коэффициента соотношения площадей.

# Найдём информацию о таунхаусах (Type), проданных компанией (SellerG) McGrath, у которых коэффициент 
# соотношения площадей здания и участка (AreaRatio) меньше -0.8. Результат отсортируем по дате продажи (Date) 
# в порядке возрастания, а после проведём сортировку по убыванию коэффициента соотношения площадей. 
# Также обновим старые индексы на новые, установив параметр ignore_index на True. Для наглядности результата 
# выберем из таблицы только столбцы SellerG и AreaRatio

mask1 = melb_df['AreaRatio'] < -0.8
mask2 = melb_df['Type'] == 'townhouse'
mask3 = melb_df['SellerG'] == 'McGrath'
melb_df[mask1 & mask2 & mask3].sort_values(
    by=['Date', 'AreaRatio'],
    ascending=[True, False],
    ignore_index=True
).loc[:, ['Date', 'AreaRatio']]
```

### Группировка данных

```python
# Применим агрегирующую функцию среднего к результату работы groupby(). В качестве столбца 
# для группировки возьмём столбец типа объекта недвижимости (Type)
melb_df.groupby(by='Type').mean()
```

### Группировка по одному признаку с несколькими агрегирующими функциями

```python
# найдём количество продаж, а также среднее и максимальное значения цен объектов недвижимости (Price), 
# сгруппированных по номеру месяца продажи (MonthSale). Результат отсортируем по количеству продаж 
# в порядке убывания:
melb_df.groupby('MonthSale')['Price'].agg(
    ['count', 'mean', 'max']
).sort_values(by='count', ascending=False)
```

```python
# Если нужна полная информация обо всех основных статистических характеристиках внутри каждой группы, 
# вы можете воспользоваться методом agg(), передав в качестве его параметра строку 'describe'
melb_df.groupby('MonthSale')['Price'].agg('describe')
```

```python
# мы можем вычислить число уникальных риелторских компаний в зависимости от региона, чтобы понять, 
# в каких регионах конкуренция на рынке недвижимости меньше. Это можно сделать, передав в параметр 
# метода agg() строку 'nunique'.
# Передадим дополнительно встроенную функцию set, чтобы получить множество из агентств недвижимости, 
# которые работают в каждом из регионов
melb_df.groupby('Regionname')['SellerG'].agg(['nunique', set])
```

### Сводные таблицы: метод pivot_table

```python
# построим ту же самую таблицу, но уже с использованием метода pivot_table
# Агрегирующую функцию оставим по умолчанию (среднее)
melb_df.pivot_table(
    values='Price',
    index='Rooms',
    columns='Type',
    fill_value=0
).round()
```

```python
# проанализируем продажи в каждом из регионов в зависимости от того, будний был день или выходной. 
# Для этого построим сводную таблицу, в которой строками будут являться названия регионов (Regionname), 
# а в столбцах будет располагаться наш «признак-мигалка» выходного дня (Weekend), который равен 1, 
# если день был выходным, и 0 — в противном случае. В качестве значений сводной таблицы возьмём количество продаж.
melb_df.pivot_table(
    values='Price',
    index='Regionname',
    columns='Weekend',
    aggfunc='count',
)
```

```python
# Найдём, как зависит средняя и медианная площадь участка (Landsize) от типа объекта (Type) 
# и его региона (Regionname). Чтобы посмотреть несколько статистических параметров, нужно передать 
# в аргумент aggfunc список из агрегирующих функций. Построим такую сводную таблицу, 
# где пропущенные значения заменим на 0

melb_df.pivot_table(
    values='Landsize',
    index='Regionname',
    columns='Type',
    aggfunc=['median', 'mean'],
    fill_value=0
)
```

### Многомерные сводные таблицы

```python
# построим таблицу, в которой по индексам будут располагаться признаки метода продажи (Method) 
# и типа объекта (Type), по столбцам — наименование региона (Regionname), а на пересечении строк 
# и столбцов будет стоять медианная цена объекта (Price)
melb_df.pivot_table(
    values='Price',
    index=['Method','Type'],
    columns='Regionname',
    aggfunc='median',
    fill_value=0
)
```

### Доступ к данным сводной таблицы

```python
# Возьмём сводную таблицу, которую мы создавали ранее
pivot = melb_df.pivot_table(
    values='Landsize',
    index='Regionname',
    columns='Type',
    aggfunc=['median', 'mean'],
    fill_value=0
)
```

```python
# Выведем её столбцы с помощью атрибута columns
pivot.columns
```

```python
# получиv средние значения площадей участков для типа здания unit, просто последовательно 
# обратившись по имени столбцов
display(pivot['mean']['unit'])
```

```python
# если нам нужны регионы, в которых средняя площадь здания для домов типа house меньше их 
# медианной площади, то мы можем найти их следующим образом:

mask = pivot['mean']['house'] < pivot['median']['house']
filtered_pivot = pivot[mask]
display(filtered_pivot)
```

```python
# Чтобы получить индексы отфильтрованной таблицы, можно воспользоваться атрибутом index 
# и обернуть результат в список:

print(list(filtered_pivot.index))
# ['Southern Metropolitan', 'Western Metropolitan']
```

### Объединение таблиц: concat
Для корректной конкатенации по строкам объединяемые таблицы должны иметь одинаковую структуру — идентичное число и имена столбцов.

```python
# склеим таблицы ratings1 и ratings2
ratings = pd.concat([ratings1, ratings2])
display(ratings)
```

```python
# по умолчанию concat сохраняет первоначальные индексы объединяемых таблиц - нужно выставить 
# параметр ignore_index на True
ratings = pd.concat(
    [ratings1, ratings2],
    ignore_index=True
)
display(ratings)
```

```python
# Давайте узнаем количество строк в таблицах ratings и dates

print('Число строк в таблице ratings: ', ratings.shape[0])
print('Число строк в таблице dates: ', dates.shape[0])
print(ratings.shape[0] == dates.shape[0])

# количество строк в таблицах разное - при разрезании таблиц какая-то строка задублировалась
```

```python
# для поиска дублированных строк выведем последнюю строку таблицы ratings1 и первую строку таблицы ratings2

display(ratings1.tail(1))
display(ratings2.head(1))
```

### Удаление дублей

```python
ratings = ratings.drop_duplicates(ignore_index=True)
print('Число строк в таблице ratings: ', ratings.shape[0])
# Число строк в таблице ratings: 100836
```

```python
# добавим к нашей таблице с оценками даты их выставления. Для этого конкатенируем таблицы ratings и dates по столбцам

ratings_dates = pd.concat([ratings, dates], axis=1)
display(ratings_dates.tail(7))
```
